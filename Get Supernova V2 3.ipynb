```python
# Get all their RA and dec

import numpy as np


RA = []
DEC = []
NAME =[]
#from     https://lweb.cfa.harvard.edu/supernova/finders/Supernovae.html
tally=0
with open("List of Supernovae.txt", encoding="utf-8") as f:
    
    # Skip header 
    for _ in range(29):
        next(f)
    
    # search each line
    for line in f:
        
        
        parts = line.split()  # splits on whitespace
        
        try:

            split_str = "    "  # search for the big space befor eht date
            # Split on the big spaces
            
            num_parts = line.split(split_str, 1)
            # parts[1] starts with the value we want
            
            # Split that part by whitespace and get first word
            after_value = num_parts[1].split()
            
            #print(after_value)
        except: 
            tally+=1
            continue
        
        
        #skip them if they are incorrectly formatted
        try:
            ss=float(after_value[0]) 
        except:
            tally+=1
            continue
            
            
        if float(after_value[0]) < 1900:
            tally+=1
            continue 
        if float(after_value[0]) < 1988:
            break
            
            
        start = parts[1].find('name="') + len('name="')
        end = parts[1].find('"', start)
        name = parts[1][start:end]
        NAME.append(name) #ID
        
        #ra and dec
        RA.append(after_value[3]+"h"+after_value[4]+"m")
        #remove plus signs from dec
        if(after_value[5][0]=="+"):
            after_value[5] = after_value[5][1:]
        if after_value[6][0]=="+":
            after_value[6] = after_value[6][1:]

        DEC.append(after_value[5]+"."+after_value[6]) 


        
        
        ###################################################
        tally+=1
        
for i in range(10000):
    print(i, NAME[i], RA[i],DEC[i])
```

```python
#save supernova coords:

with open("Supernovae_Coords.txt", "w") as f:
        
    for i in range(len(RA)):
        
        f.write(str(NAME[i]+ " " + RA[i]+" " +DEC[i]) + "\n")
        
```

```python
# save as ra dec astropy skycoord 
from astropy.coordinates import SkyCoord
import astropy.units as u

sn_coords = []

for i in range(len(RA)):
    
    ra_str, dec_str = RA[i],DEC[i]
    print("#",i,"ID",NAME[i],ra_str,dec_str)
    
    dec_deg = float(dec_str)

    
    coord = SkyCoord(ra=ra_str,dec=dec_deg * u.deg,frame='icrs')

    sn_coords.append(coord)
    
print(sn_coords[0])
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import requests


# Replace with the full path to your ChromeDriver
chromedriver_path = "/media/ben/home/ben/Documents/Grad Stuff/Supernovae/chromedriver"

service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)




radius = 1      # degrees
#tmin = 239557417 # MET start time
#tmax = 261988417 # MET end time 


tally=0




from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


for ra, dec,name in zip(RA, DEC,NAME):
    tally+=1
    
    #skips
    
    if tally<3263:
        continue
    print('tally',tally)
    # Step 1: Submit query
    params = {
        "coordfield": str(ra)+","+" "+str(dec),
        "shapefield": radius,
        "timefield": "2008-08-04 15:43:36, 2025-08-13 02:05:06",
        "energyfield": "30,1000000"    
    }
    
    
    base_query_url = "https://fermi.gsfc.nasa.gov/cgi-bin/ssc/LAT/LATDataQuery.cgi"

    driver.get(base_query_url)
    
    # Coordinates for example
    driver.find_element(By.NAME, "coordfield").send_keys(params['coordfield'])
    driver.find_element(By.NAME, "shapefield").send_keys(params['shapefield'])
    driver.find_element(By.NAME, "timefield").send_keys(params['timefield'])
    driver.find_element(By.NAME, "energyfield").send_keys(params['energyfield'])
    
    #driver.find_element(By.NAME, "Start Search").click()

    # Wait for the page to load
    time.sleep(2)

    # Locate the input by its value and click
    search_button = driver.find_element(By.XPATH, '//input[@value="Start Search"]')
    search_button.click()

    
    time.sleep(3)
    
    # Find the link and click it
    
    try:
        
        link = driver.find_element(By.XPATH, '//a[contains(@href, "https://fermi.gsfc.nasa.gov/cgi-bin/ssc/LAT/QueryResults.cgi")]')
    except:
        
        print(f"No FITS file found for RA={ra}, DEC={dec}")
        print("Quit")
        continue
     
    #####################################
    #next page
    print("Source found,",name,link.text)
    link.click()    
    
    time.sleep(3)

    driver.refresh()

    
    #####################################
    # Step 3: Download the FITS file
    
    downloaded = False
    
    # keep refreshing until the query response comes back
    while downloaded==False:
        




        links = driver.find_elements(
        By.XPATH,
        '//a[contains(@href, "https://fermi.gsfc.nasa.gov/FTP/fermi/data/lat/queries/") and contains(@href, "_PH") and contains(@href, ".fits")]'
        )


        for llll in range(len(links)):
            link=links[llll]
            print("Link href:", link.get_attribute("href"))

            # Download the file
            response = requests.get(link.get_attribute("href"))
            response.raise_for_status()


            filename = link.get_attribute("href").split("/")[-1]
            with open(filename, "wb") as f:
                f.write(response.content)

            print(f"Saved: {filename}")
        
        #print(links,len(links))
        if len(links)>0:
            downloaded=True
        else:
            #go again
            time.sleep(15)
            driver.refresh()

    
    
print("Found",tally,"sources")
driver.quit()

    
```
